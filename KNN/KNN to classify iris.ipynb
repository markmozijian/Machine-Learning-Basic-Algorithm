```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv('iris.data')
```

```python
iris.head(5)
```

```python
iris.iris_type = iris.iris_type.replace('Iris-setosa', 0)
iris.iris_type = iris.iris_type.replace('Iris-versicolor', 1)
iris.iris_type = iris.iris_type.replace('Iris-virginica', 2)
```

```python
train=iris.sample(frac=0.8,random_state=200) #random state is a seed value
test=iris.drop(train.index)
```

```python
x_train = np.array(train.iloc[:,:4])
x_test = np.array(test.iloc[:,:4])
x_train
```

```python
y_train = np.array(train.iris_type)
y_test = np.array(test.iris_type)
y_train
```

# Run knn FOR ALL test set

```python
T = 0
F = 0
k = 6
TestResult = []
for i in range(len(y_test)):
    distances = []
    for x in x_train:
        d = sqrt(np.sum((x-x_test[i])**2))
        distances.append(d)
    nearest = np.argsort(distances)
    top = [y_train[j] for j in nearest[:k]]
    y_predit = Counter(top).most_common(1)[0][0]
    print("y_predit: {}, y_real: {}".format(y_predit,y_test[i]))
    if y_predit == y_test[i]:
        T+=1
        TestResult.append(0)
    else:
        F+=1
        TestResult.append(1)
```

```python
print("True: {}, False: {}, Accurancy:{}".format(T,F,T/(T+F)))
```

```python
TestResultdf = pd.DataFrame(TestResult, columns = ['TestResult'])
TestResultdf
```

```python
TestResult = np.array(TestResultdf)
TestResult
```
